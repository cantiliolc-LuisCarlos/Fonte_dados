Análise de dados sobre sorvete

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sakshisatre/ice-cream-sales-dataset")

print("caminho para os arquivos conjuntos de dados:", path)
```

```python
import pandas as pd

df_ice_cream = pd.read_csv(f"{path}/Ice Cream.csv")
display(df_ice_cream.head(5))
```

```python
import matplotlib.pyplot as plt

# Create scatter plot
plt.scatter(df_ice_cream['Temperature'], df_ice_cream['Revenue'])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Venda')
plt.title('Venda vs Temperatura')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = df_ice_cream['Temperature'].values.reshape(-1, 1)
y = df_ice_cream['Revenue'].values

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print the model coefficients
print(f"Coeficiente angular (Slope): {model.coef_[0]}")
print(f"Coeficiente linear (Intercept): {model.intercept_}")
```
